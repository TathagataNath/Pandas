```python
import pandas as pd
```

```python
matches=pd.read_csv('matches.csv')
deliveries=pd.read_csv('deliveries.csv')
```

```python
deliveries
```

```python
#Question 1

openers=matches.drop_duplicates('season')
cities=openers.groupby('city').size()
max_host=max(cities)
for i in cities.index :
    if cities[i]==max_host :
        print(i)
```

```python
#Question 2

scoring_balls=deliveries[(deliveries['bowler']=='SL Malinga') & (deliveries['batsman_runs']>0)]
batsman=scoring_balls.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
top_3_batsman=batsman.head(3).index
for i in top_3_batsman:
    print(i)
```

```python
#Question 3

def best_bowler(team_name):
    
    wickets=deliveries[((deliveries['player_dismissed'].isnull())==False) & (deliveries['batting_team']==team_name)]
    bowlers=wickets.groupby('bowler').size().sort_values(ascending=False)
    max_wickets=max(bowlers)
    for i in bowlers.index :
        if bowlers[i]==max_wickets :
            print(i)

best_bowler('Delhi Capitals')
```

```python
#Question 4

def winning(team_name) :
    
        finals=matches.drop_duplicates('season',keep='last')
        occourance=finals[(finals['team1']==team_name) | (finals['team2']==team_name)]
        winner=occourance[occourance['winner']==team_name]
        
        return occourance.shape[0],winner.shape[0]
        
played,won=winning('Chennai Super Kings')

print("played -",played)
if won!=0 :
    print("won -",won)
```

```python
#Question 5

def score(batsman_name) :
    
    scoring_balls=deliveries[(deliveries['batsman']==batsman_name) & (deliveries['batsman_runs']>0)]
    runs=scoring_balls.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
    
    return runs

score('V Kohli')
```
