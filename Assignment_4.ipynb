```python
#Question 1

dismissals=delivery[(delivery['player_dismissed'].isnull())==False][['over','bowler','player_dismissed']]

top_10_bowlers=dismissals.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10).index.tolist()

top_10_bowlers_data=dismissals[dismissals['bowler'].isin(top_10_bowlers)]

pivot_series=top_10_bowlers_data.groupby(['over','bowler'])['player_dismissed'].count().sort_values(ascending=False)

pivot_data=pivot_series.reset_index().sort_values(['over','player_dismissed'],ascending=[True,False])

pivot_data.pivot_table(index='over',columns='bowler',values='player_dismissed',aggfunc='sum').replace(np.nan,0)
```

```python
#Question 2

def profile(player_name) :
    
    player=delivery[delivery['batsman']==player_name][['match_id','batsman_runs']]
    
    innings=player.drop_duplicates('match_id').shape[0]
    runs=player['batsman_runs'].sum()
    mom=data[data['player_of_match']==player_name].shape[0]
    boundary=player[(player['batsman_runs']==4) | (player['batsman_runs']==6)].shape[0]
    
    print("INNINGS :",innings)
    print("TOTAL RUNS :",runs)
    print("MAN OF THE MATCH :",mom)
    print("BOUNDARIES :",boundary)

profile('MS Dhoni')
```

```python
# Question 3

def top_batsman(venue_name) :
    
    all_data=data.merge(delivery,left_on='id',right_on='match_id')
    all_data=all_data[(all_data['venue']==venue_name) & (all_data['batsman_runs']>0)]
    
    return all_data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(4)

batsman=top_batsman("Wankhede Stadium")
for i in batsman.index :
    print(i,batsman[i],sep='\t')
```
